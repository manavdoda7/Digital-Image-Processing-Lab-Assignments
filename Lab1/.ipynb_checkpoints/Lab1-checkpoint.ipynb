## Task A: Reading Image

```python
img = cv2.imread('testImage.jpeg', cv2.IMREAD_UNCHANGED)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Task B: Displaying Image

```python
plt.imshow(rgb_img)
plt.show()
```

## Task C: Saving Image

```python
print('Directory before Saving the image:')
!ls
cv2.imwrite('testImage_copy.jpeg', rgb_img)
print('Directory after saving the image:')
!ls
```

## Task D: Finding Image Information

```python
img = Image.open('img.jpeg')
exifData = img.getexif()
for tagid in exifData:
    tagname = TAGS.get(tagid, tagid)
    value = exifData.get(tagid)
    print(tagname, ': ', value)
```

## Task E: Find the size of the Image

```python
dimensions = img.shape
print('Dimensions: ', dimensions)
print('Height: ', dimensions[0])
print('Width: ', dimensions[1])
print('Channels: ', dimensions[2])
```

## Task F: Resizing an Image

```python
resized_image = cv2.resize(img,(500,500)); 
cv2.imwrite('testImage_resized.jpeg', img)
print('Resized Image')
plt.imshow(resized_image)
plt.show()
```

## Task G: How to convert the color image into gray scale image.

```python
grayscaleImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print('Original Image')
plt.imshow(img)
plt.show()
print('Grayscale Image')
plt.imshow(grayscaleImage)
plt.show()
```

## Task H: Converting Grayscale Image to binary

```python
(thresh,blackAndWhiteImage) = cv2.threshold(grayscaleImage,127,255,cv2.THRESH_BINARY);
print('Grayscale Image')
plt.imshow(grayscaleImage)
plt.show()
print('Binary Image')
plt.imshow(blackAndWhiteImage)
plt.show()
```
