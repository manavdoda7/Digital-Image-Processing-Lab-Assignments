## Importing Necessary modules

```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import numpy as np
```

### Importing and Displaying Image

```python
img = cv2.imread('chessBoard.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
```

## Objective: Apply Run Length Encoding on a given Image

```python
[rows, cols] = img.shape
GL = img[0][0]
freq=0
for i in range(rows):
    for j in range(cols):
        if img[i][j] == GL:
            freq+=1
        else:
            print("(", GL, ",", freq, "); ", end = "")
            GL = img[i][j]
            freq = 1
print("(", GL, ",", freq, ")")   
```
