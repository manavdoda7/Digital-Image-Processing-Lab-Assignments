## Objective: Implement Lossy Image Compression using DPCM using Quantizer
### Importing and Displaying image

```python
img = cv2.cvtColor(cv2.imread('testImage.jpeg'), cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
plt.show()
img
```

### Compression of Image and encoding

```python
sub = 0
sub_ = 0
quantizer = 5
rows = img.shape[0]
cols = img.shape[1]
arr = []
for i in range(rows):
    for j in range(cols):
        arr.append(int((int(img[i][j])-int(sub))/quantizer))
        sub=int((int(img[i][j])-int(sub))/quantizer)*quantizer+sub
```

### Decoding and Expansion

```python
add=0
for ind in range(rows*cols):
    i = int(ind/cols)
    j = ind%cols
    img[i][j] = (arr[ind])*quantizer+add
    add=img[i][j]
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
plt.show()
img
```
